```python
import codecs
import numpy as np
import matplotlib.pyplot as plt
```

```python
log_file_baseline = '../Logs/sup.log'
total_val = 64000
num_epochs = 100

losses = np.zeros(num_epochs)
corrects = np.zeros(num_epochs)
count = 0
```

```python
with codecs.open(log_file_baseline, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if line.startswith('Valid set'):
            line = line.split()
            loss = line[-4][:-1]
            losses[count] = float(loss)
            num_corrects = line[-2].split('/')[0]
            corrects[count] = int(num_corrects)
            count = count + 1

assert count == num_epochs
```

```python
log_file_best = '../Logs/log_best.txt'
total_val = 64000
num_epochs_best = 50

loss_best = np.zeros(num_epochs_best)
correct_best = np.zeros(num_epochs_best)
count = 0
```

```python
with codecs.open(log_file_best, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if line.startswith('Valid set'):
            line = line.split()
            loss = line[-4][:-1]
            loss_best[count] = float(loss)
            num_corrects = line[-2].split('/')[0]
            correct_best[count] = int(num_corrects)
            count = count + 1
print(count)
assert count == num_epochs_best
```
